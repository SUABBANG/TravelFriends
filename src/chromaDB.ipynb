```python
# 리스트 불러오기
import pickle

with open("../data__directory/DATA/전라남도_광주_여행_contents_list.pkl","rb") as f:
    list_ex_load = pickle.load(f)
```

```python
list_ex_load
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, Document
from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import RetrievalQA, ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
```

```python
import os
api_key = os.getenv('OPENAI_API_KEY')
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002") # Embedding 모델 : text-embedding-ada-002

database = Chroma(persist_directory="../data__directory/DATA/vector_db",  
                    embedding_function = embeddings  
)
```

```python
database
```

```python
# 리스트 내용을 각각 document로 변환
documents = [Document(page_content=text) for text in list_ex_load]
```

```python
database.add_documents(documents)
```

```python
database.get()
```

```python
query = "광주에 여행 갈만한곳 추천해주세요."
```

```python
retriever = database.as_retriever()
```

```python
# model
chat = ChatOpenAI(model="gpt-4o")
qa = RetrievalQA.from_llm(llm=chat,  retriever=retriever )
```

```python
result = qa(query)
```
